```python
# ATM Withdrawal Validation System

print("===== ATM Withdrawal Validation System =====")

# User Inputs
name = input("Enter Account Holder Name: ")

balance = float(input("Enter Account Balance: "))
daily_limit = float(input("Enter Daily Withdrawal Limit: "))
withdrawn_today = float(input("Amount Withdrawn Today: "))

entered_pin = input("Enter ATM PIN: ")
withdraw_amount = float(input("Enter Withdrawal Amount: "))

# Correct ATM PIN
correct_pin = "9381"

print("\n===== Transaction Result =====")

# Validation Logic
if entered_pin == correct_pin:

    if withdraw_amount > 0:

        if withdraw_amount <= balance:

            # Check daily limit
            if (withdrawn_today + withdraw_amount) <= daily_limit:

                # Check minimum balance rule
                remaining_balance = balance - withdraw_amount

                if remaining_balance >= 1000:

                    print("Transaction Successful")
                    print("Remaining Balance:", remaining_balance)

                else:
                    print("Transaction Failed: Minimum balance rule violated")

            else:
                print("Transaction Failed: Daily limit exceeded")

        else:
            print("Transaction Failed: Insufficient balance")

    else:
        print("Transaction Failed: Invalid withdrawal amount")

else:
    print("Transaction Failed: Invalid PIN")

# Transaction Summary
print("\n===== Transaction Summary =====")
print("Account Holder Name :", name)
print("Account Balance     :", balance)
print("Daily Limit         :", daily_limit)
print("Withdrawn Today     :", withdrawn_today)
print("Withdrawal Amount   :", withdraw_amount)
```
